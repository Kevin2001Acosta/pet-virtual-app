# Tratamiento de datos

## Sesgo de neutral en los datos

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/pysentimiento_emocion.csv")
df.head()
```

```python
df.info()
```

```python
df.label.value_counts()
```

```python
# Filtrar las filas que no son 'neutral'
df_no_neutral = df[df['label'] != 'neutral']

# Seleccionar aleatoriamente 89 filas de 'neutral' (149 - 60 = 89)
df_neutral_reduced = df[df['label'] == 'neutral'].sample(n=89, random_state=42)

# Combinar ambos DataFrames
df_balanced = pd.concat([df_no_neutral, df_neutral_reduced])

# Verificar la nueva distribución de etiquetas
print(df_balanced['label'].value_counts())
```
