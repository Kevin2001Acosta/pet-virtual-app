```python
from transformers import pipeline
```

```python
model_path = "daveni/twitter-xlm-roberta-emotion-es"
emotion_analysis = pipeline("text-classification", framework="pt", model=model_path, tokenizer=model_path)
emotion_analysis("Einstein dijo: Solo hay dos cosas infinitas, el universo y los pinches anuncios de bitcoin en Twitter. Paren ya carajo aaaaaaghhgggghhh me quiero murir")
```

```python
emotion_analysis('hola')
```

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/pysentimiento_emocion.csv")
df.head()
```

```python
## analizamos cada texto en el dataframe y creamos una nueva columna con el resultado

def analizar_emocion(texto):
    result = emotion_analysis(texto)
    return 'neutral' if result[0]['label'] == 'others' else result[0]['label']

# creamos la nueva columna
df['emotion_model'] = df['transcription'].apply(analizar_emocion)
```
